```python

output_dir = "/content/drive/MyDrive/Images/results_doug"

folders = [f for f in os.listdir(output_dir) if os.path.isdir(os.path.join(output_dir, f))]
print(folders)
for folder in folders:
    folder_path = os.path.join(output_dir, folder)

    with_mask_images = sorted([os.path.join(folder_path, f) for f in os.listdir(folder_path) if f.startswith('with_mask')])
    inpainted_with_mask_images = sorted([os.path.join(folder_path, f) for f in os.listdir(folder_path) if f.startswith('inpainted_with_mask')])
    plt.figure(figsize=(10, 5 * len(with_mask_images)))

    for i, (with_mask, inpainted_with_mask) in enumerate(zip(with_mask_images, inpainted_with_mask_images)):
        with_mask_img = Image.open(with_mask)
        inpainted_with_mask_images = Image.open(inpainted_with_mask)

        plt.subplot(len(with_mask_images), 2, 2 * i + 1)
        plt.imshow(with_mask_img)
        plt.axis('off')
        plt.title(f"{folder} - With Mask {i}")

        plt.subplot(len(with_mask_images), 2, 2 * i + 2)
        plt.imshow(inpainted_with_mask_images)
        plt.axis('off')
        plt.title(f"{folder} - Filled With Mask {i}")

    plt.tight_layout()
    plt.savefig(f"content/drive/MyDrive/Images/plots/plot_{folder}.png")
    plt.show()
```
